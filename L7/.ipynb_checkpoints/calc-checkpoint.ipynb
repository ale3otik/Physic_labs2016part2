```python
U = np.array([10.3,9.27,8.18,7.14,6.18,5.15,4.25,3.2,2.3,1.2]) * 10
Xm = np.array([3.8,3.4,3,2.6,2.2,1.9,3* 0.5,2.3* 0.5,1.7 * 0.5, 2.2 * 0.2])
Ym = np.array([3* 5,2.8* 5,2.8* 5,2.6* 5,2* 5,0.8*5,1.4,2.5*0.2,2.6*0.1,2.3*0.05]) 
Dm = Ym
Em = Xm
```

```python
plt.figure(figsize=(15,7))
plt.grid(True)
plt.title(r'$D_m(E_m)$',fontsize=20)
plt.xlabel(r'$E_m$ [N/ql]',fontsize=20)
plt.ylabel(r'$D_m$ [N/ql]',fontsize=20)
plt.plot(Em,Dm,'-o',color='black')
plt.savefig('plot1.png')
plt.show()
```

```python
eps = Dm/Em
plt.figure(figsize=(15,7))
plt.grid(True)
plt.title(r'$\varepsilon(E_m)$',fontsize=20)
plt.xlabel(r'$E_m$ [N/ql]',fontsize=20)
plt.ylabel(r'$\varepsilon$',fontsize=20)
plt.plot(Em,eps,'-o',color='black')
plt.savefig('plot2.png')
plt.show()
```
